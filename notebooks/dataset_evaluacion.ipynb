analisis de correlacion: Ausencia contra notas

```python
# --- PREPARACIÓN DE ASISTENCIA ---
# Limpiamos el estado de asistencia
df_asistencia['estado_asistencia'] = df_asistencia['estado_asistencia'].str.strip().str.upper()

# Contamos cuántas "AUSENTE" tiene cada inscripción
# (Cada fila en asistencia es una clase)
ausencias_por_inscripcion = df_asistencia[df_asistencia['estado_asistencia'] == 'AUSENTE'].groupby('id_inscripcion').size().reset_index(name='total_ausencias')

# --- PREPARACIÓN DE NOTAS ---
# Limpiamos notas (como hicimos antes) y filtramos outliers
df_calificaciones['nota'] = pd.to_numeric(df_calificaciones['nota'], errors='coerce')
df_calif_limpio = df_calificaciones[(df_calificaciones['nota'] >= 1.0) & (df_calificaciones['nota'] <= 7.0)]

# Sacamos el promedio de notas por cada inscripción
notas_por_inscripcion = df_calif_limpio.groupby('id_inscripcion')['nota'].mean().reset_index(name='promedio_notas')

# --- EL CRUCE (JOIN) ---
# Unimos ambos datos por el id_inscripcion
df_correlacion = pd.merge(ausencias_por_inscripcion, notas_por_inscripcion, on='id_inscripcion')

# --- CÁLCULO DE CORRELACIÓN ---
corr_valor = df_correlacion['total_ausencias'].corr(df_correlacion['promedio_notas'])

print(f"La correlación entre Ausencias y Notas es: {corr_valor:.2f}")

# Visualización 

plt.figure(figsize=(10, 6))
sns.regplot(data=df_correlacion, x='total_ausencias', y='promedio_notas', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title(f'Relación entre Ausencias y Rendimiento Académico (Corr: {corr_valor:.2f})')
plt.xlabel('Cantidad de Ausencias')
plt.ylabel('Promedio de Notas')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

0.27 se ve reflejado solo por el outlier

```python
# Veamos los valores máximos de ambas columnas en nuestra tabla cruzada
display(df_correlacion.describe().round(2))

# Miremos directamente quién es el que tiene más ausencias y más nota
print("\n--- Top 3 con más Ausencias ---")
display(df_correlacion.sort_values(by='total_ausencias', ascending=False).head(3))

print("\n--- Top 3 con mayor Promedio ---")
display(df_correlacion.sort_values(by='promedio_notas', ascending=False).head(3))
```

```python
# --- LA FORMA CORRECTA DE CONTAR AUSENCIAS (INCLUYENDO LOS 0) ---
# Creamos una columna matemática: Vale 1 si faltó, vale 0 si vino
df_asistencia['es_ausente'] = (df_asistencia['estado_asistencia'] == 'AUSENTE').astype(int)

# Sumamos esa columna por cada estudiante. ¡Los perfectos tendrán un 0!
ausencias_reales = df_asistencia.groupby('id_inscripcion')['es_ausente'].sum().reset_index(name='total_ausencias')

# --- EL CRUCE REAL ---
# Unimos todos (los que faltaron y los de asistencia perfecta) con sus notas
df_correlacion_real = pd.merge(ausencias_reales, notas_por_inscripcion, on='id_inscripcion')

# --- RECALCULAMOS ---
corr_real = df_correlacion_real['total_ausencias'].corr(df_correlacion_real['promedio_notas'])

print(f"La correlación REAL (incluyendo 0 faltas) es: {corr_real:.2f}")

# El gráfico definitivo
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(data=df_correlacion_real, x='total_ausencias', y='promedio_notas', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title(f'Relación Real: Ausencias vs Notas (Corr: {corr_real:.2f})')
plt.xlabel('Cantidad de Ausencias (Incluye asistencia perfecta)')
plt.ylabel('Promedio de Notas')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

no hay correlacion en alumnos que faltan y sus notas
